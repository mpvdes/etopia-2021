## Longitud y Latitud
Vamos a hacer dos listas una de longitudes y otra de latitudes
Explicar:
 * bucles
 * añadir a lista

```python
df['geometry']
```

```python
l = []
l = l + [1,2]
l = l + [3]
l
```

```python
longitudes = []
for coordenadasStr in df['geometry']:
    pointStr = coordenadasStr.split(',')
    longitudes += [float(pointStr[0])]
```

```python
latitudes = []
for coordenadasStr in df['geometry']:
    pointStr = coordenadasStr.split(',')
    latitudes += [float(pointStr[1])]
```

```python
coordenadas = pd.DataFrame({'long':longitudes,'lat':latitudes})
coordenadas
```

```python
df
```
